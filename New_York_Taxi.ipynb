# Dataset Name - New York City Taxi Trips 

### Extracting Json Data from MongoDB and Storing it in a Data Frame - 'df_nyc_taxi'

```python
#importing library 
import pymongo
import pandas as pd
from pandas import DataFrame

#Connecting to local MongoDB database 
client = pymongo.MongoClient("mongodb://localhost:27017/")

# Database Name
db_name = client["DAP_Project"]

# Collection Name
collection = db_name["New_York_Taxi"]

#Extarcting and Storing the data in the Data Frame
df_nyc_taxi = DataFrame(list(collection.find({})))
df_nyc_taxi
```

```python
row_and_columns = list(df_nyc_taxi.shape)


print("\nNumber of Rows in the Dataset : ", row_and_columns[0])
print("Number of Columns in the Dataset : ", row_and_columns[1])
```

##### Exploring the Columns/Features of the Dataset

```python
column_names = list(df_nyc_taxi.columns)

print("\nThe Dataset contains following Columns :")
for column in column_names:
    print(column)
 
print("\nStatistics of the Data is as follows :")
df_nyc_taxi.describe()
```

##### Checking the datatypes of the Columns

```python
print("\nDatatypes of the Columns")
print('--------------------------------')
print("\nColumns with 'int' datatype : ")
for i in list(df_nyc_taxi.select_dtypes(include=['int64']).columns):
    print(i)

print('\n--------------------------------')
print("Columns with 'float' datatype : ")
for f in list(df_nyc_taxi.select_dtypes(include=['float64']).columns):
    print(f)

print('\n--------------------------------')
print("Columns with 'String' datatype : ")
for s in list(df_nyc_taxi.select_dtypes(include=['object']).columns):
    print(s)
```

As seen the above results, the columns - **pickup_datetime** and  **dropoff_datetime** falls under the data-type of *string*, which technically should fall under the *datetime*.

Also all the instances in these two columns have a string 'T' in them, which should be removed first removed before converting the data type to datetime

##### Removing the string ' T '   from the instances of column - 'pickup_datetime' and 'dropoff_datetime' and changing their data-type to ' datetime ' 

```python
# Removing 'T' from the columns - 'pickup_datetime' and 'dropoff_datetime'
df_nyc_taxi['pickup_datetime'] = df_nyc_taxi['pickup_datetime'].replace("T", " ", regex=True)
df_nyc_taxi['dropoff_datetime'] = df_nyc_taxi['dropoff_datetime'].replace("T", " ", regex=True)

df_nyc_taxi.head()
```

```python
#Changing the data-type of 'pickup_datetime' and 'dropoff_datetime' to datetime

df_nyc_taxi['pickup_datetime'] = pd.to_datetime(df_nyc_taxi['pickup_datetime'], format='%Y-%m-%d %H:%M:%S.%f')
df_nyc_taxi['dropoff_datetime'] = pd.to_datetime(df_nyc_taxi['dropoff_datetime'], format='%Y-%m-%d %H:%M:%S.%f')
```

```python
#Checking if the data-type as changed
print('pickup_datetime data-type : ', df_nyc_taxi['pickup_datetime'].dtype)
print('dropoff_datetime data-type : ', df_nyc_taxi['dropoff_datetime'].dtype)
```

##### Investigating for null values

```python
print('Columns and their number of null-values : ')
df_nyc_taxi.shape
for na in df_nyc_taxi.columns:
    print(na,' - ',df_nyc_taxi[na].isna().sum())
```

All the values in the column **' rate_code '** are **Null**. Similarly for the column **' store_and_fwd_flag '** , 999938 instances out of 1000000 have a **Null value**. So it is better to drop these two columns

In addition to that, the  **'_id '** column in the Dataset holds unique ID's of each row. Since, the ' _id ' column does not hold much relevance in the analysis, it is better to drop it.

```python
#Dropping the '_id' column
df_nyc_taxi = df_nyc_taxi.drop(['_id'], axis = 1)
print(" \n '_id' Column dropped from the dataset")
df_nyc_taxi = df_nyc_taxi.drop(['rate_code'], axis = 1)
print(" \n 'rate_code' Column dropped from the dataset")
df_nyc_taxi = df_nyc_taxi.drop(['store_and_fwd_flag'], axis = 1)
print(" \n 'store_and_fwd_flag' Column dropped from the dataset")

print('\n Number of Columns reaming : ', list(df_nyc_taxi.shape)[1])
df_nyc_taxi
```

Now, the dataset is free from NULL value entries and the data-type of each of the columns is lined as per the feature. 

This *cleaned*, *processed* and *structured* version of the dataset will now be stored in the ***PostreSQL database***, which can be retrived later for further analysis. 


### Storing the Processed and Structured Data in PostgreSQL 

```python
#importing the libraries
from sqlalchemy import create_engine
import psycopg2
import pandas as pd

#Establishing the connection with the local PostreSQL database
connection = create_engine('postgresql://postgres:admin@localhost:5432/dap')

#Sending and Storing the data in the PostgreSQL database
#The name of the table where the dataset will be stored is - NewYorkTaxiTrips
df_nyc_taxi.to_sql('NewYorkTaxiTrips',connection)
```
